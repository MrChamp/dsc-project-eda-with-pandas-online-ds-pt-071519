## Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centers.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.

```python
#centrality/dispersion:
crime = df['crim']
tax = df['tax']
ptratio = df['ptratio']

crime.plot.hist(bins = 30)
print(crime.describe())
print('median   ', crime.median())
#visually the crime variable is clearly skewed which is backed up by noticing
#the mean and median differ by about 171%
```

```python
#if the data is split into the highly dense bin vs the rest it might be interesting to see what the data looks like:
#ill split the data points in half after sorting them and the first half of the data should be more representative of the 
#entire data set than the second half

sorted_crime = crime.sort_values()
crime_a = sorted_crime.iloc[:166]
crime_b = sorted_crime.iloc[167:]
```

```python
crime_a.plot.hist(bins = 50)
print(crime_a.describe())
print('median   ', crime_a.median())
```

```python
crime_b.plot.hist(bins = 50)
print(crime_b.describe())
print('median   ', crime_b.median())
```

```python
tax.plot.hist(bins = 30)
print(tax.describe())
print('median   ', tax.median())
#Also a skewed variable, full-value property-taxe rate per $10,000 has about a
#21% difference between the mean and median. There also seems to be a distinct
#gap in values for the data, which might be worth investigating by splitting
#the data around that gap
```

```python
sorted_tax = tax.sort_values()
tax_a = sorted_tax.iloc[:240]
tax_b = sorted_tax.iloc[241:]
# tax_b.head(75)
```

```python
tax_a.plot.hist(bins = 50)
print(tax_a.describe())
print('median   ', tax_a.median())
#tax_a gives a distribution much closer to a normal. After dropping the outliers (values>=666) this data seems representative
#of an entire subgroup of full-value property-taxe rates
```

```python
tax_b.plot.hist(bins = 50)
print(tax_b.describe())
print('median   ', tax_b.median())

#tax_b describes the set of the outlier data and has enough values to be representative of something, but is different
#enough from the values in tax_a that some useful information may be extracted by separating them. in fact almost all the 
#values for this data subset appear to be located at 666
```

```python
ptratio.plot.hist(bins = 30)
print(ptratio.describe())
print('median   ', ptratio.median())
#with about a 3 percent difference between the mean and the median, this data
#could be reasonabally summarized with either measure. 
```
